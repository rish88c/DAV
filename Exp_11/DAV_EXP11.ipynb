```r
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)A
import matplotlib.pyplot as plt
import seaborn as sns  # visualization tool
```

```r
data = pd.read_csv("NBA_player_of_the_week.csv")
```

```r
data.head()
```

```r
data.info()
```

```r
data.dropna(inplace=True)
```

```r
data.drop(['Real_value'],axis=1,inplace=True)
```

```r
# Correlation's heatmap.
f,ax = plt.subplots(figsize=(10,8))
sns.heatmap(data.corr(),annot=True, linewidths=.5,fmt=".1f",ax=ax)
plt.show()
```

```r
plt.figure(figsize=(8, 6))
sns.countplot(x='Position', data=data, palette='coolwarm')
plt.title('Distribution of Player Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

```r
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of NBA Players')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```r
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Height', y='Weight', data=data, hue='Position', palette='Set1')
plt.title('Height vs. Weight of NBA Players')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend(loc='upper left')
plt.show()
```

```r
plt.figure(figsize=(12, 8))
top_teams = data['Team'].value_counts().head(10)
sns.barplot(x=top_teams.values, y=top_teams.index, palette='viridis')
plt.title('Top 10 Teams with Most Player of the Week Awards')
plt.xlabel('Number of Awards')
plt.ylabel('Team')
plt.show()
```

```r
# Load required libraries
library(ggplot2)
library(reshape2)
```

```r
data <- read.csv("NBA_player_of_the_week.csv")
```

```r
# Display the first few rows of the data
head(data)
```

```r
# Display information about the data
str(data)
```

```r
# Drop rows with missing values
data <- na.omit(data)
```

```r
# Drop the column 'Real_value'
data <- subset(data, select = -c(Real_value))

# Correlation heatmap
correlation_matrix <- cor(data[, sapply(data, is.numeric)])
heatmap(correlation_matrix,
        Rowv = NA, Colv = NA,
        col = colorRampPalette(c("darkblue", "white", "darkred"))(100),
        scale = "none", margins = c(5, 5))
```

```r
# Distribution of Player Positions
ggplot(data, aes(x = Position, fill = Position)) +
  geom_bar() +
  ggtitle("Distribution of Player Positions") +
  xlab("Position") +
  ylab("Count") +
  theme_minimal()
```

```r
# Age Distribution of NBA Players
ggplot(data, aes(x = Age)) +
  geom_histogram(binwidth = 1, fill = "skyblue", color = "black") +
  ggtitle("Age Distribution of NBA Players") +
  xlab("Age") +
  ylab("Count") +
  theme_minimal()
```

```r
# Height vs. Weight of NBA Players
ggplot(data, aes(x = Height, y = Weight, color = Position)) +
  geom_point() +
  ggtitle("Height vs. Weight of NBA Players") +
  xlab("Height") +
  ylab("Weight") +
  theme_minimal()
```

```r
# Top 10 Teams with Most Player of the Week Awards
top_teams <- head(sort(table(data$Team), decreasing = TRUE), 10)
top_teams_df <- data.frame(Team = names(top_teams), Awards = as.numeric(top_teams))
ggplot(top_teams_df, aes(x = reorder(Team, -Awards), y = Awards)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  ggtitle("Top 10 Teams with Most Player of the Week Awards") +
  xlab("Team") +
  ylab("Number of Awards") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
